```python
final_df = pd.merge(df,df1,on="Country Code", how="left")
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
### pie chart - top 3 countries uses zomato

plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

obervation: zomato is mostly used in india and then in united states and united kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

### obsevation

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating_count",hue="Rating color",data=ratings)
```

```python
### COUNT PLOT
sns.countplot(x="Rating color",data=ratings)
```
